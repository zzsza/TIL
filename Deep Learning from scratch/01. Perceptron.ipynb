```python
import numpy as np
```

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(1, 1)
```

```python
AND(1, 0)
```

```python
def AND2(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND2(1,1)
```

```python
AND2(0,1)
```

```python
AND2(0,0)
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x)+b
    if tmp < 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(0,1)
```

```python
NAND(1,1)
```

```python
NAND(1,0)
```

```python
def OR(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(0,1)
```

```python
OR(1,0)
```

```python
OR(1,1)
```
