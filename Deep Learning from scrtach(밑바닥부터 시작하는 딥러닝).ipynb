### 밑바닥부터 시작하는 딥러닝 - 이론은 대부분 아는 내용이기에, 직접 구현을 해보고 빠르게 넘어가기!

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(1, 1)
```

```python
AND(1, 0)
```

```python
def AND2(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND2(1,1)
```

```python
AND2(0,1)
```

```python
AND2(0,0)
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x)+b
    if tmp < 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(0,1)
```

```python
NAND(1,1)
```

```python
NAND(1,0)
```

```python
def OR(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(0,1)
```

```python
OR(1,0)
```

```python
OR(1,1)
```

```python
# xor 문제는 퍼셉트론으로 풀 수 없다.
# 그러나 직선이라는 제약을 없애면 가능! -> 다층 퍼셉트론을 사용
```

```python
def XOR(x1, x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

# 신경망
- 활성화함수 : 입력 신호의 총합을 출력 신호로 변환하는 함수
- 퍼셉트론은 계단함수를 활성화 함수로 사용!


```python
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

```python
np.arange?
```

```python
x = np.arange(-5.0, 5.0, 0.1)
# 기존엔 np.arange([-5.0, 5.0, 0.1])로 했으나 [] 없이 사용 가능하게 변함
y = step_function(x)
plt.ylim(-0.1, 1.1)
plt.plot(x, y)
plt.show()
```

## 시그모이드 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-5.0, 5.0, 0.1])
```

```python
sigmoid(x)
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

- 선형 함수를 사용하면 신경망의 층을 깊게 하는 의미가 없어지기에 비선형 함수를 활성화 함수로 사용!
- 선형 함수는 상수배만큼 변하는 함수라, 은닉층이 없는 네트워크로 표현 가능함.

## ReLU 함수
- 입력이 0을 넘으면 그 입력을 그대로 출력하고, 0 이하면 0을 출력하는 함수

```python
def relu(x):
    return np.maximum(0, x)
```
